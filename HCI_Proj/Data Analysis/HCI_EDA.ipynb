```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pylab as plt
import scipy.stats as st
import pingouin as pg

from statsmodels.stats.anova import AnovaRM
from statsmodels.stats.multicomp import pairwise_tukeyhsd

%matplotlib inline
```

```python
data = pd.read_csv('test_data.csv')
data 
```

```python
df = data.copy()
df_grouped = df.groupby(by=['Subject','Keyboard_layout', 'Input_method'])[['WPM', 'TotErrRate']].mean().reset_index()
df_grouped
```

```python
for depvar in ['WPM', 'TotErrRate']:
    
    stat, p = st.levene(df_grouped[depvar][df['Keyboard_layout'] == 'QUADRANT LAYOUT'],
                           df_grouped[depvar][df['Keyboard_layout'] == 'QWERTY LAYOUT'])
    print('Levene\'s Test: Statistics=%.3f, p=%.3f' % (stat, p),'for', depvar)
```

```python
for depvar in ['WPM','TotErrRate']:
# Peform the repeated measures ANOVA
    anova = AnovaRM(df_grouped, depvar=depvar, subject='Subject', within=['Keyboard_layout', 'Input_method'])
    res = anova.fit()
    
    print(depvar,'\nRepeated Measures',res,'\n')
```

```python
for dep_var in ['WPM','TotErrRate']:
    for indep_var in ['Input_method', 'Keyboard_layout']:
        # Compare 'WPM' means across different 'Typing_method' using Tukey's HSD test
        tukey = pairwise_tukeyhsd(endog=df_grouped[dep_var],     # Data
                                  groups=df_grouped[indep_var],   # Groups
                                  alpha=0.05)          # Significance level
        
        print(dep_var,'Tukey\'s HSD test regarding',indep_var,':\n',tukey,'\n\n')
```

```python
unique_KL = df_grouped.Keyboard_layout.unique().tolist()
unique_IM = df_grouped.Input_method.unique().tolist()
```

```python
selection = []
features_combination = []
for KL in unique_KL:
    for IM in unique_IM:
        sel = (df_grouped.WPM.where((df_grouped['Input_method'] == IM) &
                                      (df_grouped['Keyboard_layout'] == KL)).to_frame().dropna().WPM.tolist())
        features_combination.append([KL, IM])
        selection.append(sel)


selection
```

```python
features_combination
```

```python
plt.figure(figsize=(8,8))
finalWPM = pd.DataFrame()
name = []
for i, fc in enumerate(features_combination):
    n = features_combination[i][0]+' + '+features_combination[i][1]
    name.append(n)
    finalWPM[n] = selection[i]

ax=sns.barplot(selection)
xtick_loc = ax.get_xticks()
xtick_labels = name

# Set the x-axis ticks
ax.set_xticks(ticks = xtick_loc, labels = xtick_labels, rotation = 45, ha = 'right')
ax.set(ylabel = 'WPM mean', xlabel = 'type of keyboard + input method', title = 'WPM mean over all combinations of input and keyboard layout')
```

```python
df = data.copy()
wpm_err_max = pd.DataFrame(df.groupby(by=['Trial','Keyboard_layout', 'Input_method'])[['WPM', 'TotErrRate']].max().reset_index()[['WPM', 'TotErrRate']])
wpm_err_min = pd.DataFrame(df.groupby(by=['Trial','Keyboard_layout', 'Input_method'])[['WPM', 'TotErrRate']].min().reset_index()[['WPM', 'TotErrRate']])
wpm_err_min = wpm_err_min.rename(columns={'WPM':'WPM_min', 'TotErrRate':'TotErrRate_min'})
```

```python
wpm_err_max = wpm_err_max.rename(columns={'WPM':'WPM_max', 'TotErrRate':'TotErrRate_max'})
```

```python
result = pd.concat([df_grouped2,wpm_err_min, wpm_err_max],axis=1)
result
```

```python
trials = result.groupby(by='Trial')[['WPM', 'TotErrRate', 'WPM_min', 'TotErrRate_min','WPM_max','TotErrRate_max']].mean()
trials
```

```python
x = [1,2,3]
plt.figure(figsize=(8,8))
plt.xticks([1,2,3])
plt.title('Graph of WPM (mean, min, max) per trial across all tests')
plt.ylabel('WPM')
plt.xlabel('Trial number')
plt.plot(x, trials.WPM, marker='o', label='mean WPM per trial', color='b')
plt.plot(x, trials.WPM_max, marker='^', label='max WPM per trial', color='r')
plt.plot(x, trials.WPM_min, marker='v', label='min WPM per trial', color='g')
plt.legend(loc='best')
plt.show()
```

```python
x = [1,2,3]
plt.figure(figsize=(8,8))
plt.xticks([1,2,3])
plt.title('Graph of total error rate (mean, min, max) per trial across all tests')
plt.ylabel('Total error rate')
plt.xlabel('Trial number')
plt.plot(x, trials.TotErrRate, marker='o', label='mean total error rate per trial', color='b')
plt.plot(x, trials.TotErrRate_max, marker='^', label='max total error rate per trial', color='r')
plt.plot(x, trials.TotErrRate_min, marker='v', label='min total error rate per trial', color='g')
plt.legend(loc='best')
plt.show()
```
